```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
csvFile = open("PracticalSalinityUnits_Dec_2011.csv", 'r') 
csvReader = csv.reader(csvFile, delimiter=',')
data_lists = list(csvReader)
```

```python
width = len(data_lists[0])
height = len(data_lists)
```

```python
grid = np.empty([height, width, 3], dtype=np.uint8) 
```

```python
all_values = [] 
for r, row in enumerate(data_lists):
    for c, value in enumerate(row):
        if float(value) < 99999:
               all_values.append(float(value)) # This is where we fill in the list with all data values
        if float(value) <= 30: 
               grid[r, c] = [2, 59, 247]
        elif float(value) <= 31: 
               grid[r, c] = [1, 242, 247]
        elif float(value) <= 32: 
               grid[r, c] = [121, 16, 150]
        elif float(value) <= 33: 
               grid[r, c] = [1, 58, 16]
        elif float(value) <= 34: 
               grid[r, c] = [49, 224, 96]
        elif float(value) <= 35: 
               grid[r, c] = [238, 242, 4]
        elif float(value) <= 36: 
               grid[r, c] = [239, 138, 7]
        elif float(value) <= 37: 
               grid[r, c] = [244, 73, 73]
        elif float(value) <= 38: 
               grid[r, c] = [173, 0, 0]
        else:
               grid[r, c] = [0, 0, 0]
```

```python
print(min(all_values))
print(max(all_values))
```

```python
plt.title("December 2011 Sanility Levels")
plt.imshow(grid)
plt.axis('off')
#plt.savefig("Land_temperature.png",dpi = 1000)
plt.show()
```

```python
import pandas as pd 
```

```python
orig_df = pd.read_csv("Salinity_info_38N_68W.csv")
orig_df.head()
df = orig_df.dropna()
```

```python
depth = df["Depth "]
temp = df["Temperature (Celsius)"]
sal = df["Salinity (PSU)"]
den= df["Density (kg/m3)"]

print(max(depth))
print(min(depth))
print(max(temp))
print(min(temp))
print(max(sal))
print(min(sal))
print(max(den))
print(min(den))
```

```python
plt.clf()
plt.plot(depth, temp, color = "teal")
plt.twinx()
plt.plot(depth, den, color = "pink")
plt.ylabel("Temperature and Density of the Ocean in various Depths")
plt.xticks(range(0, 4001, 500))
plt.xlim(0, 4000)
plt.savefig("Temp and Den")
plt.show()
```

```python
plt.clf()
plt.plot(depth, temp, color = "teal")
plt.legend()
plt.plot(depth, sal, color = "purple")
plt.legend()
plt.title("Temperature and Salinity of the Ocean in various Depths")
plt.ylim(0, 40)
plt.yticks(range(0, 40, 4))
plt.xlim(0, 4001)
plt.xlabel("Depth")
plt.savefig("Temp and sal")
plt.show()
```
